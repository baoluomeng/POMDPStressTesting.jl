```julia
using MCTS
using POMDPs
using POMDPModels # for the GridWorld problem
using D3Trees
using Random
mdp = SimpleGridWorld()
solver = MCTSSolver(n_iterations=1000, depth=20, exploration_constant=5.0, enable_tree_vis=true)
planner = solve(solver, mdp)

s = initialstate(mdp, Random.MersenneTwister(4))
a = action(planner, s)
```

```julia
a, info = action_info(planner, s)
D3Tree(info[:tree], init_expand=2)
```
